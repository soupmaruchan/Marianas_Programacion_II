## 1. Operadores en Python

Los operadores son símbolos o palabras especiales q permiten realizar
operaciones sobrevariables y valores.

En Python se clasifican en:

- Aritméticos
- De comparación
- Lógicos
- De asignación
- De pertenencia
- De identidad

```python
# Ejemplo de operadores aritméticos

a = 15
b = 4

print("Suma:", a + b)          # Suma dos números
print("Resta:", a - b)         # Resta
print("Multiplicación:", a * b)
print("División:", a / b)
print("División entera:", a // b)
print("Residuo:", a % b)
print("Potencia:", a ** b)
```

### Operadores de Asignación

Se utilizan para asignar valores a variables.

=, +=, -=, *=, /=, //=, %=, **=

```python
x = 10
x += 5
print(x)
```

in, not in

```python
a = 5
b = 5
print(a is b)
```

## 2. Palabras Reservadas

Las palabras reservadas son términos que tienen un significado especial
en Python y no pueden utilisarse como identificadores.

```python
import keyword

print("Lista oficial de palabras reservadas en Python:")
print(keyword.kwlist)
```

Los operadores permiten manipular datos y realizar cálculos,
