# Introducción

Python es un lenguaje de programación que utiliza operadores y palabras reservadas para realizar operaciones matemáticas, comparaciones lógicas y controlar el flujo del programa, y pues comprender estos elementos es esencial para escribir código correcto y estructurado.

# Operadores en Python

## Operadores Aritméticos

Permiten realizar operaciones matemáticas.

| Operador | Significado | Ejemplo |
|----------|------------|----------|
| + | Suma | 10 + 5 |
| - | Resta | 10 - 5 |
| * | Multiplicación | 10 * 5 |
| / | División | 10 / 5 |
| // | División entera | 10 // 3 |
| % | Módulo | 10 % 3 |
| ** | Potencia | 2 ** 3 |

## Operadores de Comparación

Devuelven valores booleanos (True o False).

| Operador | Significado |
|----------|------------|
| == | Igual que |
| != | Diferente de |
| > | Mayor que |
| < | Menor que |
| >= | Mayor o igual |
| <= | Menor o igual |

## Operadores Lógicos

Se utilizan para combinar condiciones.

- **and** → Verdadero si ambas condiciones se cumplen  
- **or** → Verdadero si al menos una se cumple  
- **not** → Invierte el valor lógico  

## Operadores de Asignación

Se utilizan para asignar valores a variables.

- =  
- +=  
- -=  
- *=  
- /=  

## Operadores de Membresía

- in  
- not in  

## Operadores de Identidad

- is  
- is not  

# Palabras Reservadas en Python

Las palabras reservadas tienen un significado especial y no pueden usarse como nombres de variables.


```python
# Ejemplo usando operadores y palabras reservadas

x = 10
y = 5

if x > y and x != 0:
    print("x es mayor que y")

for i in range(3):
    print("Iteración:", i)
```

## Conclusión

Los operadores permiten realizar cálculos y evaluaciones lógicas, mientras que las palabras reservadas estructuran el comportamiento del programa.  
Conocer estos elementos es fundamental para desarrollar programas en Python de manera correcta y eficiente.
